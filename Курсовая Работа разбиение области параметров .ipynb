<h1>Разбиение области параметров для первых 8 элементов </h1>

Тк как для delta=1.4 S для четных элементов имеет отрицательное значение , я решил в силу инвариантности заменф (γ, φ, y) ⇒ (−γ, −φ, −y) использовать модуль S

re-пересчитывает гамма , skoaf- вычисляет коэффициент S для данного номера элемента 

```python
import math 
def re(k):
    z=[]
    oblastgran=[1,1,.64,.64,.4,.4,.21,.21]
    delta=-0.7
    for i in range(0,len(b)):
        newgamma=b[i]*abs(skoaf(k))
        ##print(newgamma,"-для ",k)
        ##print(oblastgran[k-1])
        if newgamma>=oblastgran[k-1]:
            z.append(oblastgran[k-1])
        else:    
            z.append(newgamma)  
    return z  
def skoaf(k):
    return (1+delta)/(1-(-delta)**k)
```
